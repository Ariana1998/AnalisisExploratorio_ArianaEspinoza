## Notebook Pima Indian Diabetes Dataset

Este conjunto de datos se encarga de mostrarnos un análisis de la tribu Pima respecto al porcentaje de diabetes que padecen las mujeres, ya sea a causa de los embarazos o que tengan cierta predisposición a padecerlo, a continuación se explicara el análisis de los respectivos datos.

Primero se importan las librerías necesarias para utilizar este conjunto de datos.

```python
import numpy as np # Librería de Python especializada en el cálculo numérico y el análisis de datos.
import pandas as pd # Librería de Python especializada en la manipulación y el análisis de datos.
import matplotlib as mpl # Biblioteca para la generación de gráficos en dos dimensiones.
import matplotlib.pyplot as plt # Módulo Matplotlib que propone varias funciones para añadir elementos como líneas, imágenes o textos a los ejes de un gráfico.
import seaborn as sns # Biblioteca para crear gráficos estadísticos. 
```

Posteriormente se importan los datos en un dataframe para poder manipularlos y ejemplificarlos con gráficas.

```python
df = pd.read_csv(r'diabetes.csv')
```

Se imprime el conjunto de datos para tenerlo de forma visual en una lista.

```python
df
```

Hacemos un histograma de cada uno de los datos estudiados en el conjunto de datos: Número de embarazos, glucosa, presión arterial, grosor de la piel, insulina, indice de masa corporal, predisposición a la diabetes, edad y si padecen diabetes o no.

```python
df.hist()
plt.figure(figsize=(10,5))
```

En este conjunto de datos se puede apreciar que las mujeres en su mayoría están entre sus 20 y 30 años al momento de recabar los datos.

```python
df['Age'].value_counts().plot(kind="density")
```

Podemos apreciar que la gran mayoría de las mujeres ha tenido un bebé, en el siguiente lugar no han tenido ninguno o han tenido 2 hasta el momento.

```python
df['Pregnancies'].value_counts().plot(kind="pie")
```

Se puede observar que la mayorías de las mujeres de la tribu Pima no padecen diabetes, pero aproximadamente un 33% si lo padece.

```python
df['Outcome'].value_counts().plot(kind="pie")
```

En esta gráfica podemos ver el nivel de glucosa de las mujeres examinadas con respecto a su edad, viendo así que las mujeres que tienen aproximadamente 22 tuvieron el nivel más alto en glucosa.

```python
plt.style.use('bmh')
x = df['Age']
y = df['Glucose']

#Bar chart
plt.xlabel('Edad')
plt.ylabel('Nivel de Glucosa')
plt.bar(x,y)

```

En esta gráfica podemos ver que la mayoría de las mujeres en este conjunto de datos tienen un índice de masa corporal de 30-34, teniendo así obesidad de grado 1, mientras que en el siguiente lugar con mayor cantidad de personas tienen un índice de masa corporal de 35-39 el cual es obesidad tipo 2. Dándonos cuenta así que en la tribu pima las mujeres en su mayoría tienen obesidad tanto grado 1 como grado 2.

```python
df['BMI'].plot.hist(edgecolor="white")
```

En la gráfica podemos ver por la cantidad de niños que han tenido las mujeres en la tribu y la respectiva edad actual de las mismas.

```python
Embarazos = df['Age']
Edad = df['Pregnancies']
fig = plt.figure()
Myaxes = fig.add_axes([0.1, 0.1, 1.6, 1.6])
x_ticks = np.arange(0,18)
plt.xticks(x_ticks)
Myaxes.bar(Embarazos,Edad)
```

Aquí se ejemplifica el número de embarazos con relación al estado de la presión arterial, ya que en algunas ocasiones el embarazo deja como secuela tener que tratarse la hipertensión de por vida.

```python
# make data:
x = df['Pregnancies']
y = df['BloodPressure']


# plot
fig, ax = plt.subplots()
ax.bar(x, y, width=1, edgecolor="white", linewidth=0.7)
plt.xlabel('Embarazos')
plt.ylabel('Presion Arterial')
plt.show()
```

En la gráfica se muestran los embarazos en relación a la predisposición genética a padecer diabetes, ya sea por la edad y sus antecedentes familiares.

```python
# make data:
x = df['Age']
y = df['DiabetesPedigreeFunction']

# plot
fig, ax = plt.subplots()
ax.bar(x, y, width=1, edgecolor="white", linewidth=0.7)
plt.xlabel('Edad')
plt.ylabel('Predisposicion diabetica')
plt.show()
```

Aquí se muestra el número de embarazos respecto al nivel de insulina en las mujeres, debido a que en algunas ocasiones el cuerpo no produce la suficiente insulina, provocando la diabetes gestacional. Normalmente esta desaparece después del parto, pero en algunas ocasiones no desaparece y se transforma en diabetes tipo 2.

```python
# make data:
x = df['Pregnancies']
y = df['Insulin']

# plot
fig, ax = plt.subplots()
ax.bar(x, y, width=1, edgecolor="white", linewidth=0.7)
plt.show()
```

En conclusión, se puede decir que las mujeres de la tribu Pima tienen un alto índice de obesidad grado 1 y 2, ya sea por consecuencia del embarazo o su predisposición genética, además de que la gran mayoría ha tenido mínimo un bebe. Este conjunto de datos bien podría utilizarse para realizar un estudio y ver si es viable proveerle a estas mujeres un control de su salud de forma personalizada, para así evitar que caigan en algún grado de obesidad, tengan problemas de glucosa o hipertensión, pero principalmente que todas estas problemáticas no acaben en un cuadro de diabetes. 
En el momento que se recabaron los datos existía un porcentaje de 33% de mujeres padeciendo esta enfermedad, pero no sabemos si a futuro aumentara el porcentaje, por eso sería buena idea fomentar el cuidado de la salud para que el porcentaje vaya disminuyendo con el paso del tiempo.
